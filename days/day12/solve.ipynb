```python
DIRECTIONS = [(-1,0), #up
              (0,1),  #right
              (1,0),
              (0,-1)]


def create_regions(garden_map: list):
    regions = np.zeros(input.shape, dtype=int)
    n_regions = 0

    for  i in range(garden_map.shape[0]):
        for j in range(garden_map.shape[1]):
            # check up and left
            val = garden_map[i,j]
            if i > 0 and j > 0:
                above_me = garden_map[i-1,j]
                left_me = garden_map[i,j-1]
                if val == above_me and val == left_me:
                    if regions[i-1,j] == regions[i,j-1]:
                        regions[i,j] = regions[i-1,j]
                    else:
                        min_region = min(regions[i-1,j], regions[i,j-1])
                        max_region = max(regions[i-1,j], regions[i,j-1])
                        regions[regions == max_region] = min_region
                        regions[i,j] = min_region
                    continue
            if i > 0:
                above_me = garden_map[i-1,j]
                if val == above_me: 
                    regions[i,j] = regions[i-1,j]
                    continue
            if j > 0:
                left_me = garden_map[i,j-1]
                if val == left_me: 
                    regions[i,j] = regions[i,j-1]
                    continue
            regions[i,j] = int(n_regions)
            n_regions += 1
    return regions, n_regions
            
regions, n_regions = create_regions(input)
n_regions, regions
```

```python
def area(region_idx: int, regions):
    return int(np.sum(regions == region_idx))

def perimeter(region_idx: int, regions):
    positions = np.argwhere(regions == region_idx)
    per = 0
    for pos in positions:
        per += perimeter_plot_part(tuple(pos), regions)
    return per

def perimeter_plot_part(pos, regions):
    i,j = pos
    per = 0
    for dir in DIRECTIONS:
        new_pos = (i+ dir[0], j+dir[1])
        # check outside (+1)
        if not ( (0 <= new_pos[0] < regions.shape[0]) and (0 <= new_pos[1] < regions.shape[1]) ):
            per += 1
        elif regions[pos] != regions[new_pos]:
            per += 1
    return per
        
def price(regions):
    n_regions = np.max(regions) + 1
    price = []
    for region_idx in range(n_regions):
        price += [area(region_idx, regions) * perimeter(region_idx, regions)]
    print(price)
    return sum(price)
area(2, regions), perimeter(2, regions)
price(regions)
```

```python
def fences(region_idx, regions):
    fences_horz = np.zeros((regions.shape[0]+1, regions.shape[1]), dtype=int)
    fences_vert = np.zeros((regions.shape[0], regions.shape[1]+1), dtype=int)

    positions = np.argwhere(regions == region_idx)
    for pos in positions:
        pos = tuple(pos)
        i,j = pos
        for dir in DIRECTIONS:
            is_fence = False
            new_pos = (i+ dir[0], j+dir[1])
            if not (0 <= new_pos[0] < regions.shape[0]) or not (0 <= new_pos[1] < regions.shape[1]):
                is_fence = True
            elif regions[pos] != regions[new_pos]:
                is_fence = True
            if is_fence:
                if dir[0] == 0:
                    fences_vert[i,j+(1+dir[1])//2] = dir[1]
                if dir[1] == 0:
                    fences_horz[i+(1+dir[0])//2,j] = dir[0]
    return fences_horz, fences_vert

def perimeter_count_fences(region_idx, regions):
    fences_horz, fences_vert = fences(region_idx, regions)
    num_sides_horz, num_sides_vert = 0, 0

    for i in range(fences_horz.shape[0]):
        for j in range(fences_horz.shape[1]):
            if fences_horz[i,j] != 0:
                if  j==0 or (j!=0 and (fences_horz[i,j-1] != fences_horz[i,j])):
                    num_sides_horz +=1
    
    for j in range(fences_vert.shape[1]):
        for i in range(fences_vert.shape[0]):
            if fences_vert[i,j] != 0:
                if i==0 or (i!=0 and (fences_vert[i-1,j] != fences_vert[i,j])):
                    num_sides_vert +=1
    
    per = num_sides_horz + num_sides_vert
    return per


def price_sides(regions):
    n_regions = np.max(regions) + 1
    price = []
    for region_idx in range(n_regions):
        price += [area(region_idx, regions) * perimeter_count_fences(region_idx, regions)]
    print(price)
    return sum(price)


price_sides(regions)
```
