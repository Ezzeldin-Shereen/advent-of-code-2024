```python
from itertools import product
possible_ops = ["+", "*"]

def interleave_eq_eval(vals: list, ops: list):
    # vals has one more element than ops
    res = vals[0]
    for val, op in zip(vals[1:], ops):
        eq = f"{res}{op}{val}"
        res = eval(eq)
    return res

def is_calibration_possible(line: tuple, ops: list):
    result, values = line
    num_ops = len(values)-1
    for op_combination in product(ops, repeat=num_ops):
        res = interleave_eq_eval(values, op_combination)
        if res == result:
            return True
    return False
```

```python
possible_res = [line[0] for line in lines if is_calibration_possible(line, possible_ops)]
sum(possible_res)
```

```python
possible_ops2 = possible_ops + [""]
possible_res = []
for i, line in enumerate(lines):
    print(f"Parsing line {i+1} of {len(lines)}")
    if is_calibration_possible(line, possible_ops2):
        possible_res.append(line[0])
sum(possible_res)
```
