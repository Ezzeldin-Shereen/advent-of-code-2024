```python
import numpy as np

with open("input.txt", 'r') as f:
    input = f.read()

SIZE = 71
CAP = 1024

# SIZE = 7
# CAP = 12
input1 = """5,4
4,2
4,5
3,0
2,1
6,3
2,4
1,5
0,6
3,3
2,6
5,1
1,2
5,5
2,5
6,5
1,4
0,4
6,4
1,1
6,1
1,0
0,5
1,6
2,0"""

corrupted = [tuple(coord.split(",")) for coord in input.split("\n")]
corrupted = [(int(t[0]), int(t[1])) for t in corrupted]
corrupted_subset = corrupted[:CAP]
corrupted_subset
```

```python
DIRECTIONS = [(-1,0),
              (0,1),
              (1,0),
              (0,-1)]
START = (0,0)
END = (SIZE-1, SIZE-1)

def explore(start, corrupted):
    costs = np.ones((SIZE,SIZE), dtype=float)*float("inf")
    dirs = np.zeros((SIZE,SIZE))
    costs[start] = 0
    to_explore = [(start)]
    while len(to_explore) > 0:
        pos = to_explore[0]
        for dir in DIRECTIONS:
            next_pos = pos[0]+dir[0], pos[1]+dir[1]
            # check whether next position is corrupted or outside memory
            if (next_pos not in corrupted) and (next_pos[0] in range(SIZE)) and (next_pos[1] in range(SIZE)):
                current_cost = costs[pos]+1                    
                if current_cost < costs[next_pos]:
                    costs[next_pos] = current_cost
                    dirs[next_pos]
                    # since we updated the cost, we should explore again
                    to_explore.append(next_pos)
        to_explore = to_explore[1:]
    return costs

costs = explore(START, corrupted_subset)
int(costs[END])
```
