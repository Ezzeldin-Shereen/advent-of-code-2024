## Imports

```python
import numpy as np
```

## Read

```python
with open('input.txt', mode='r') as f:
    lines = f.readlines()

lines[-1]
```

create 2 lists

```python
l1 = [int(line.split("   ")[0].strip()) for line in lines]
l2 = [int(line.split("   ")[1].strip()) for line in lines]

l1[-1], l2[-1]
```

reorder

```python
l1_sorted = sorted(l1)
l2_sorted = sorted(l2)
```

compute

```python
sum([abs(int(i1) - int(i2)) for (i1, i2) in zip(l1_sorted, l2_sorted)])
```

### Part 2

```python
arr1 = np.array(l1)
arr2 = np.array(l2)
times = np.array([int(sum(arr2[arr2 == i1])) for i1 in arr1])
times
```

```python
result = int(sum(times))
result
```
